```python
# 进行ocr
import fitz  # PyMuPDF
import os
import pandas as pd
import numpy as np
from tqdm import tqdm
import json
base_dir = '/data/coding/patent_qa/test/'
pdf_file_list = [x for x in os.listdir(base_dir+'/pdf_img/')]
# 前400个，排序
#pdf_file_list = sorted(pdf_file_list)[:400]

query_list = []
response_list = []
document_list = []
page_num_list = []
img_path_list = []
prompt="你是一个OCR专家，请提取以下图片中的所有文字内容，并原样返回。<image>"
for pdf_file in pdf_file_list:
    file_name = pdf_file.split('.')[0]
    file_list = [x for x in os.listdir(base_dir+'/pdf_img/'+file_name) if 'jpg' in x] # pdf文件存储的jpg
    for k in tqdm(range(len(file_list))):
        img_path = base_dir+'/pdf_img/'+file_name + '/' + file_list[k]
        page_num = int(file_list[k].split('.')[0])
        document_list.append(file_name)
        page_num_list.append(page_num)
        img_path_list.append(img_path)
        query_list.append(prompt)

# 生成jsonl文件
# with open("test_ocr_image_swift_in.jsonl", "w", encoding="utf-8") as f:
#     for i in range(len(document_list)):
#         # 读取图片并转换为base64
#         json_line = {
#             "query": query_list[i],
#             "response": str(i),
#             "pdf_name": document_list[i],
#             "pdf_page": str(page_num_list[i]),
#             "images": [img_path_list[i]],
#         }
#         # 写入文件
#         f.write(json.dumps(json_line, ensure_ascii=False) + "\n")

for i in range(len(document_list)):
    # 读取图片并转换为base64
    json_line = {
        "query": query_list[i],
        "response": str(i),
        "pdf_name": document_list[i],
        "pdf_page": str(page_num_list[i]),
        "images": [str(img_path_list[i])],
    }
    # 写入jsonl
    with open(f"test_ocr_image_swift_input_{i//400}.jsonl", "a", encoding="utf-8") as f:
        f.write(json.dumps(json_line, ensure_ascii=False) + "\n")
```
